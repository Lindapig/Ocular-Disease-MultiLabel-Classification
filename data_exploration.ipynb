```python
import pandas as pd
import matplotlib.pyplot as plt
from visualization import visualize_batch
from data_loader import create_datasets 
from config import IMAGE_INFO
from utils import read_file
```

```python
df = read_file(IMAGE_INFO)
df.keys()
selected_df = df[['N', 'D', 'G','C', 'A', 'H', 'M', 'O']]
```

```python
for category_column in selected_df.keys():
    category_distribution = selected_df[category_column].value_counts()
    print(f"\n{category_column} Distribution of categories:")
    num_0,num_1 = category_distribution.values
    distribution_0,distribution_1 = num_0/(num_0+num_1),num_1/(num_0+num_1)
    print(distribution_0,distribution_1)
```
